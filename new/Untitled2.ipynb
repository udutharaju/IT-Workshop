```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(0,3*3.24,0.1)
y_s = np.sin(x)
y_c = np.cos(x)
plt.subplot(2, 1, 1)


plt.plot(x, y_s)
plt.title('Sine')
plt.subplot(2, 1, 2)
plt.plot(x, y_c)
plt.title('Cosine')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)

plt.plot(x, y_sin)
plt.title('Sine')
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')
plt.show()
```
