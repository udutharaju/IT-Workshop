```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)

plt.subplot(2,1,1)
plt.legend(title ="Data 1 ,Data 2")

plt.scatter(x,y,color = "blue",linewidth = 0.1,label="data 1")
x = np.arange(0,3*np.pi,0.2)
y = np.cos(x)
plt.subplot(2,1,1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.scatter(x,y,color = "red",linestyle = "dashed",marker="s",label="data 2")
plt.title("scatter plot of Data")
plt.legend(loc="upper right")

plt.show()

```

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)
y_cos = np.cos(x)



plt.scatter(x,y,color = "blue",linewidth = 0.1,label="data 1")
plt.scatter(x,y_cos,color = "red",linestyle = "dashed",marker="s",label="data 2")


plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot of Data")
plt.legend(loc="upper right")

plt.show()

```

```python
import matplotlib.image as mpimg
import numpy as np
# Load an image
image = mpimg.imread('')
# Create some data to plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots()
# Display the image as the background
ax.imshow(image, extent=[0, 10, -1, 1])
# Plot the data
ax.plot(x, y, 'r-', linewidth=2)
# Adjust the limits and add labels
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```
