```python
from matplotlib import pyplot as plt

x=[12,10,4,3,2]
y=[20,8,9,3,1]


plt.plot(x,y,color="blue",linestyle="dashed")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.title("plot chart")

plt.show()
```

```python
from matplotlib import pyplot as plt

x=[1,2,4,3,5]
y=[10,20,30,40,50]



plt.xlabel("students")
plt.ylabel("marks")
c=["green","blue","yellow","pink","orange"]
plt.barh(x,y,color=c,height=0.5)
plt.title("Bar chart")

plt.show()
```

```python
from matplotlib import pyplot as plt

x=[1,2,4,3,5]
y=[10,20,30,40,50]



plt.xlabel("students")
plt.ylabel("marks")
c=["green","blue","yellow","pink","orange"]
plt.bar(x,y,color=c,width=0.5)
plt.title("Bar chart")

plt.show()
```

pie chart

```python
from matplotlib import pyplot as plt
rgukt=["cse","ece","mech","chemical"]
count=[270,170,120,80]
c=["green","yellow","blue","red"]
plt.pie(count,colors=c,explode=[0,0,0,0.3],shadow=True,labels=rgukt,autopct="%.2f%%")
plt.legend(rgukt)
plt.title("representing branch count")

plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,24,3,9]
y=[15,4,13,9]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("subplot")

x1=[14,4,13,19]
y1=[10,2,3,8]
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.title("subplot")

plt.show()
```
